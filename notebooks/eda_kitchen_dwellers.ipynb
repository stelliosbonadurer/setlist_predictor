```python
import pandas as pd
from pathlib import Path

data_dir = Path.cwd() / "data"
if not data_dir.exists():  # if the kernel started inside notebooks/
    data_dir = Path.cwd().parent / "data"

df = pd.read_csv(data_dir / "kitchen_dwellers_setlists.csv")
df.head()

df.columns = df.columns.str.strip().str.rstrip(',')
```

```python
print("Number of shows:", df['show_id'].nunique())
```

```python
df['song_name'].value_counts().head(90)
```

```python
import matplotlib.pyplot as plt

top_songs = df["song_name"].value_counts().head(20)

plt.figure(figsize=(12, 6))
top_songs.plot(kind="bar")
plt.title("Top 20 Most Played Songs  Kitchen Dwellers")
plt.ylabel("Count")
plt.xlabel("Song")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
